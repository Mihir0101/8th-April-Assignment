```python
import pandas as pd
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['area_type','availability'],axis=1,inplace=True)
```

```python
df.drop('society',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['size'] = df['size'].str.split(' ').str[0].astype(float)
```

```python
df.dropna(0,inplace=True)
```

```python
df_location_stat = df.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
df_location_lessthan_10 = df_location_stat[df_location_stat < 10]
```

```python
df['location'] = df['location'].apply(lambda x : 'Other' if x in df_location_lessthan_10 else x)
```

```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df[~df['total_sqft'].apply(is_float)]
```

```python
def normalize_sqft(x) :
  varr = x.split('-')
  if len(varr) == 2:
    return (float(varr[0]) + float(varr[1])) / 2

  try :
    return float(x)
  except :
    return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(normalize_sqft)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
location_encoded = encoder.fit_transform(df[['location']]).toarray()
```

```python
df_location_encoded = pd.DataFrame(location_encoded,columns=encoder.get_feature_names_out())
```

```python
final_df = pd.concat([df,df_location_encoded],axis=1)
```

```python
final_df.drop('location',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Price'] = df['price']
```

```python
final_df.drop('price',axis=1,inplace=True)
```

```python
final_df.dropna(0,inplace=True)
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
params = {'kernel':['linear','poly'],
          'degree':[1,3],
          'epsilon':[1,3]}
```

```python
hyper = GridSearchCV(svr,param_grid=params,cv=5)
```

```python
hyper.fit(x_train,y_train)
```

```python
hyper.best_params_
```

```python
svr = SVR(degree=1,epsilon=3,kernel='linear')
```

```python
svr.fit(x_train,y_train)
```

```python
y_predicted = svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print('Accuracy of the model is {accuracy}'.format(accuracy=r2_score(y_test,y_predicted)))
```
